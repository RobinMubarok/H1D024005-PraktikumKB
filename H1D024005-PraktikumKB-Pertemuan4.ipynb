```python
# DATABASE: Kerusakan Laptop
kemungkinan_kerusakan = {
  "RAM rusak": {"blue screen", "lag", "gagal booting"},
  "PSU lemah": {"performa turun", "bau terbakar", "kipas berisik"},
  "Overheat": {"performa turun", "panas", "kipas berisik"},
  "VGA bermasalah" : {"artefak visual", "layar berkedip", "driver error"},
  "Hardisk corrupt" : {"bunyi aneh", "file hilang", "gagal booting"}
}
#catat gejala
kerusakan = []
#tanya gejala
def tanya(gejala_kunci, pertanyaan):
  jawaban = input(f"{pertanyaan} (y/t): ").lower()
  if jawaban == "y":
    kerusakan.append(gejala_kunci)
#beri hasil
def hasil_diagnosa():
  hasil = []
  gejala_rusak = set(kerusakan)
  for rusak, gejala in kemungkinan_kerusakan.items():
    if gejala.issubset(gejala_rusak):
      hasil.append(rusak)
  return hasil if hasil else ["Tidak terdeteksi kerusakan"]
#sarankan
def saran(hasil):
  print("\nKerusakan yang Laptop Anda alami:")
  for rusak in hasil:
    print("\t", rusak)
  print("Yang Anda bisa lakukan:")
  for rusak in hasil:
    if rusak == "RAM rusak":
      print("\tCoba bersihkan atau pindah slot RAM")
    elif rusak == "PSU lemah":
      print("\tBersihkan saja debu dan udara pada PSU")
    elif rusak == "Overheat":
      print("\tGanti pasta")
    elif rusak == "VGA bermasalah":
      print("\tBersihkan VGA atau cek BIOS/UEFI")
    elif rusak == "Hardisk corrupt":
      print("\tInstal ulang driver")
    else:
      print("\tPindah servis tempat lain saja")
def main():
  print("=== SISTEM PAKAR DIAGNOSA MALARIA ===")
  print("Jawablah pertanyaan berikut dengan 'y' untuk Ya atau 't' untuk Tidak.\n")
  tanya("blue screen", "Laptop Anda menampilkan Blue Screen?")
  tanya("lag", "Apa laptop Anda nge-lag?")
  tanya("gagal booting", "Apakah laptop tidak bisa menyala normal?")
  tanya("performa turun", "Laptop Anda tidak lancar saat menjalankan aplikasi?")
  tanya("bau terbakar", "Laptop Anda bau terbakar?")
  tanya("kipas berisik", "Kipas laptop Anda berisik?")
  tanya("panas", "Laptop Anda panas?")
  tanya("artefak visual", "Layar laptop Anda berbintik?")
  tanya("layar berkedip", "Layar laptop Anda berkedip?")
  tanya("driver error", "Ada notifikasi driver error?")
  tanya("bunyi aneh", "Laptop Anda berbunyi \'ngiiing\'?")
  tanya("file hilang", "File Anda aman?")
  saran(hasil_diagnosa())
main()
```
