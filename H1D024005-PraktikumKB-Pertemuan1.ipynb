```python
import ctypes as cici
import decimal as desi

barang = {
    'id_barang' : [cici.c_int(1), cici.c_int(2), cici.c_int(3)],
    'nama_barang' : ['Apel', 'Durian', 'Pensil'],
    'harga' : [desi.Decimal(1.0), desi.Decimal(2.75), desi.Decimal(1.25)]
}

pembeli = {
    'id_pembeli' : [cici.c_int(1), cici.c_int(2), cici.c_int(3)],
    'nama_pembeli' : ['Adi', 'Shelly', 'Nisa']
}

transaksi = {
    'id_transaksi' : [cici.c_int(1), cici.c_int(2)],
    'id_pembeli' :[cici.pointer(pembeli['id_pembeli'][1]), cici.pointer(pembeli['id_pembeli'][0])],
    'id_barang' : [cici.pointer(barang['id_barang'][0]), cici.pointer(barang['id_barang'][2])]
}
```

```python
# select
for id, nama, harga in zip(barang['id_barang'], barang['nama_barang'], barang['harga']):
  print(f'{id.value}\t|\t{nama}\t|\t{harga}')
```

```python
# update
transaksi['id_barang'][1].contents.value = 7
if barang['id_barang'][2].value == 3:
  print("Belum Berubah")
else:
  print("Berubah")
```
